## 1️⃣ Verify GPU (T4)

```python
import subprocess
result = subprocess.run(['nvidia-smi'], capture_output=True, text=True)
print(result.stdout)
if 'T4' in result.stdout:
    print('✅ Tesla T4 GPU confirmed!')
elif result.returncode != 0:
    print('❌ No GPU found. Go to Runtime → Change runtime type → GPU → T4 → Save.')
else:
    print('⚠️  GPU found but not T4 — check runtime type.')
```

## 3️⃣ Create Project Folder Structure in Google Drive

```python
from pathlib import Path

# Root of the project inside Google Drive
DRIVE_ROOT = Path('/content/drive/MyDrive/SatelliteInterpolation')

folders = [
    'data/raw',
    'data/processed',
    'data/train',
    'data/validation',
    'data/test',
    'outputs/baseline',
    'outputs/finetuned',
    'outputs/metrics',
    'outputs/heatmaps',
    'outputs/animations',
    'outputs/comparisons',
    'checkpoints',
    'logs',
    'models',
]

for folder in folders:
    path = DRIVE_ROOT / folder
    path.mkdir(parents=True, exist_ok=True)

print('✅ Drive folder structure created:')
for f in sorted(folders):
    print(f'   Drive/SatelliteInterpolation/{f}')
```
